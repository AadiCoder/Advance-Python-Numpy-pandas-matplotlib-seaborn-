```python
df=pd.read_csv("company.csv")
```

```python
df
```

```python
tp=df.iloc[:,8:]
tu=df.iloc[:,7:8]
toothp=df.iloc[:,3:4]
fc=df.iloc[:,1:2]
fw=df.iloc[:,2:3]
mn =df.iloc[:,0:1]
```

```python
tp.isnull().sum()
```

```python
tu.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(tu,tp,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
lnr=LinearRegression()

lnr.fit(xtrain,ytrain)
```

```python
predicted_price=lnr.predict(xtest) 
predicted_price
```

```python
print(xtest)
print(ytest)
```

```python
plt.scatter(xtest,ytest,color='r')
plt.plot(xtest,predicted_price,color='g')
plt.title("Linear Regression")
plt.xlabel("Total unit")
plt.ylabel("Total price")
```

```python
plt.plot(mn,tp,color='g')
plt.xlabel("Month no.")
plt.ylabel("Total profit")
plt.title('Company profit per month')
```

```python
plt.plot(mn, fc,   label = 'Face cream Sales Data')
plt.plot(mn, fw,   label = 'Face Wash Sales Data')
plt.plot(mn, toothp, label = 'Tooth Paste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title('Sales data')
```
