# Merge

```python
data1 = {
  "name": ["Shally", "Kritika", "Amisha"],
  "age": [22, 24, 22]
}

data2 = {
  "name": ["Shally", "Kritika", "Manisha"],
  "age": [77, 24, 22]
}

```

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
newdf = pd.merge(df1,df2,how="right")
newdf1 = pd.merge(df1,df2,how="inner")
newdf2 = pd.merge(df1,df2,how="left")
newdf3 = pd.merge(df1,df2,how="outer")
```
